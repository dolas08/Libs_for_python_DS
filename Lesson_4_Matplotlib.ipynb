Создайте массивы y1, y2, y3, y4 по следующим формулам:

y1 = x**2

y2 = 2 * x + 0.5

y3 = -3 * x - 1.5

y4 = sin(x)

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов.

В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
# создание массивов x и y
x = np.linspace(-3,3,51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
# создание графиков 
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x,y1)
ax1.set_title('График y1')
ax1.set_xlim([-5,5])

ax2.plot(x,y2)
ax2.set_title('График y2')

ax3.plot(x,y3)
ax3.set_title('График y3')

ax4.plot(x,y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

Приравняйте к нулю отрицательные числа в матрице D_inv, а положительные - к единице. 

Убедитесь, что в матрице D_inv остались только нули и единицы.

С помощью функции numpy.where, используя матрицу D_inv в качестве маски, а матрицы B и C - в качестве источников данных, получите матрицу E размером 4x4.

Элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 1, должны быть равны соответствующему элементу матрицы B, а элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 0, должны быть равны соответствующему элементу матрицы C.

```python
D_inv_plus = D_inv > 0
D_inv_minus = D_inv < 0
```

```python
D_inv[D_inv_plus] = 1
D_inv[D_inv_minus] = 0
D_inv
```

```python
E = np.where(D_inv, B, C)
E.shape
```

```python
D_inv = D_inv.astype(int)

print('E[D_inv_plus] == B[D_inv_plus] \n', E[D_inv_plus] == B[D_inv_plus])
print('E[D_inv_plus] == C[D_inv_plus] \n', E[D_inv_minus] == C[D_inv_minus])
```
